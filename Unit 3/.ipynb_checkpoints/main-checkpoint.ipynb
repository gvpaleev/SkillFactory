Очищяем от лишних символов, заполняем пропуски, формируем dummy признаки ...

Заполняем пропуски тремя самыми популярными кухнями

```python
arr=[]

#соберем все кухни в одно ...
for row in df["Cuisine Style"]:
    if type(row)!=type('str'):
        continue 
    for item in row.split(','):
        arr.append(item)
        
        
        
df["Cuisine Style"] = df["Cuisine Style"].fillna( str(list(pd.Series(arr)\
                                     .value_counts()[0:3]\
                                     .index))\
#берем 3 самых популярные кухни и убераем из str лишнее
                           .replace('[','')\
                           .replace(']','')\
                           .replace(' ','')\
                           .replace("'",'')\
                           .replace('"','')
                          )
                                   
```

Заменим на dummy 

```python
df['Price Range'] = df['Price Range'].fillna('$$ - $$$')
```

```python
df = pd.concat ([df, pd.DataFrame(columns=['minMoney','meanMoney','maxMoney']) ] )
```

```python
df[df.columns[-3:]] = df[df.columns[-3:]].fillna(0)
```

```python
for index, row in df.iterrows():
    if (row["Price Range"] == '$$ - $$$'):
        df.loc[index,'meanMoney']=1
    if (row["Price Range"] == '$'):
        df.loc[index,'minMoney']=1
    if (row["Price Range"] == '$$$$'):
        df.loc[index,'maxMoney']=1
```

```python
X = df.drop('Rating',axis = 1 )
y = df['Rating']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25)
```

```python
from sklearn.ensemble import RandomForestRegressor # инструмент для создания и обучения модели
from sklearn import metrics # инструменты для оценки точности модели
```

```python
regr = RandomForestRegressor(n_estimators=100)
```

```python
regr.fit(X_train,y_train)
```

```python
y_preg = regr.predict(X_test)
```

```python
y_preg
```

```python
print('MAE:',metrics.mean_absolute_error(y_test,y_preg))
```
